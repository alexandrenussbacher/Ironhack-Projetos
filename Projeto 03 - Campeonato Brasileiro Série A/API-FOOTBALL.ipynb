# API-FOOTBALL

```python
import requests
import pandas as pd
import dotenv
import os
```

```python
dotenv.load_dotenv("headers.env")
```

```python
x=os.getenv("x-rapidapi-key")
y=os.getenv("x-rapidapi-host")
headers = {"x-rapidapi-key": x, "x-rapidapi-host": y}
```

```python
def brasil_2019_ligas():
    url_ligas="https://api-football-v1.p.rapidapi.com/v2/leagues/country/brazil/2019"
    response=requests.get(url_ligas, headers=headers)
    dicionario=pd.json_normalize(response.json())["api.leagues"][0]
    ligas_brasil=pd.DataFrame(dicionario)
    return ligas_brasil
```

```python
id_ligas_2019=brasil_2019_ligas()[["league_id", "name"]]
id_ligas_2019
```

```python
id_ligas_2019={"Serie A": 357, "Serie B": 358}
```

```python
def times_2019():
    df=pd.DataFrame()
    for i in [id_ligas_2019["Serie A"], id_ligas_2019["Serie B"]]:
        url_times=f"https://api-football-v1.p.rapidapi.com/v2/teams/league/{i}"
        response=requests.get(url_times, headers=headers)
        dicionario=pd.json_normalize(response.json())["api.teams"][0]
        times_serieA=pd.DataFrame(dicionario)
        df_temp=pd.DataFrame(dicionario)
        df=df.append(df_temp)
    return df
```

```python
id_times_2019=times_2019().reset_index().loc[:, ["team_id", "name"]]
id_times_2019
```

```python
id_times_2019["team_id"].values
```

```python
def jogadores_2019():
    df=pd.DataFrame()
    for i in id_times_2019["team_id"].values:
        url_jogadores=f"https://api-football-v1.p.rapidapi.com/v2/players/team/{i}"
        response=requests.get(url_jogadores, headers=headers)
        dicionario=pd.json_normalize(response.json())["api.players"][0]
        df_temp=pd.DataFrame(dicionario)
        df=df.append(df_temp)
    return df
```

```python
data_jogadores_2019=jogadores_2019()
data_jogadores_2019
```

```python
pd.options.display.max_columns=100
```

```python
jogadores_2019_geral=data_jogadores_2019.loc[((data_jogadores_2019["league"]=="Serie A") | (data_jogadores_2019["league"]=="Serie B"))].reset_index()
jogadores_2019=jogadores_2019_geral.loc[jogadores_2019_geral["season"]=="2019"].reset_index()
jogadores_2019.head()
```

```python
jogadores_2019=jogadores_2019.loc[:, ["player_id", "player_name", "position", "age", "birth_date", "nationality", "team_name", "league", "shots", "goals", "passes", "tackles", "dribbles", "fouls", "games"]]
jogadores_2019
```

```python
jogadores_2019.columns=["id", "nome", "posição", "idade", "data_nascimento", "nacionalidade", "time", "liga", "chutes", "gols1", "passes", "desarmes1", "dribles", "faltas", "jogos"]
```

```python
a=jogadores_2019.merge(pd.json_normalize(jogadores_2019["chutes"]), on=jogadores_2019.index).drop("key_0", 1)
a.rename(columns={"total": "total_chutes", "on": "chutes_no_gol"}, inplace=True)
b=a.merge(pd.json_normalize(a["gols1"]), on=a.index).drop("key_0", axis=1)
b.rename(columns={"total": "gols", "assists": "assistências", "conceded": "gols_sofridos", "saves": "defesas"}, inplace=True)
c=b.merge(pd.json_normalize(b["passes"]), on=b.index).drop(["key_0"], axis=1)
c.rename(columns={"total": "total_passes", "key": "passes_chaves", "accuracy": "%passes_certos"}, inplace=True)
d=c.merge(pd.json_normalize(c["desarmes1"]), on=c.index).drop(["key_0", "blocks", "interceptions"], axis=1)
d.rename(columns={"total": "desarmes"}, inplace=True)
e=d.merge(pd.json_normalize(d["dribles"]), on=d.index).drop(["key_0"], axis=1)
e.rename(columns={"attempts": "dribles_tentados", "success": "dribles_realizados"}, inplace=True)
f=e.merge(pd.json_normalize(e["faltas"]), on=e.index).drop(["key_0"], axis=1)
f.rename(columns={"drawn": "faltas_sofridas", "committed": "faltas_cometidas"}, inplace=True)
jogadores_2019_ok=f.merge(pd.json_normalize(f["jogos"]), on=f.index).drop(["key_0"], axis=1)
jogadores_2019_ok.rename(columns={"appearences": "total_jogos", "minutes_played": "minutos_jogados", "lineups": "jogos_escalados"}, inplace=True)
```
